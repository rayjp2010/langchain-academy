```python
from dotenv import load_dotenv
from langchain_google_genai import ChatGoogleGenerativeAI

load_dotenv()


llm = ChatGoogleGenerativeAI(model="gemini-2.5-flash", temperature=0)
```

```python
def book_hotel(hotel_name: str) -> str:
    """Book a hotel"""
    return f"Booked hotel: {hotel_name}"

def book_flight(flight_number: str) -> str:
    """Book a flight"""
    return f"Booked flight: {flight_number}"

def book_car(car_name: str) -> str:
    """Book a car"""
    return f"Booked car: {car_name}"
```

```python
from langgraph.prebuilt import create_react_agent
from langgraph_supervisor import create_supervisor

flight_assistant = create_react_agent(
    model=llm,
    tools=[book_flight],
    prompt="You are a travel assistant. You are given a task to book a flight. You can use the tools provided to you to book the flight.",
    name="flight_assistant",
)

hotel_assistant = create_react_agent(
    model=llm,
    tools=[book_hotel],
    prompt="You are a travel assistant. You are given a task to book a hotel. You can use the tools provided to you to book the hotel.",
    name="hotel_assistant",
)

car_assistant = create_react_agent(
    model=llm,
    tools=[book_car],
    prompt="You are a travel assistant. You are given a task to book a car. You can use the tools provided to you to book the car.",
    name="car_assistant",
)


supervisor = create_supervisor(
    model=llm,
    agents=[flight_assistant, hotel_assistant, car_assistant],
    prompt="""
    You are a travel assistant. You manage the agents to book a hotel, flight, and car when planning a trip.
    When you plan the trip, you should just book some random hotel, flight, and car.
    """,
    name="supervisor",
).compile()

for event in supervisor.stream({"messages": "Book Sheraton hotel in Tokyo"}, stream_mode="updates"):
    print(event)
```
